```python
from transformers import AutoModelForImageClassification, AutoProcessor
from PIL import Image
import torch

# Load the model and processor
model_name = "andupets/real-estate-image-classification"
model = AutoModelForImageClassification.from_pretrained(model_name)
processor = AutoProcessor.from_pretrained(model_name)

# Load the image
image_path = "data-scraping/watermarked_images/watermarked_image_0.jpg"
image = Image.open(image_path).convert("RGB")

# Preprocess the image
inputs = processor(images=image, return_tensors="pt")

# Make predictions
with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits

# Get the predicted class
predicted_class_idx = logits.argmax(-1).item()
print(f"Predicted class index: {predicted_class_idx}")
```

```python
from ultralytics import YOLO
from PIL import Image

# Load the YOLOv8 model
model = YOLO('fraud-filter-system/copyright-image-detection/watermark-detection/watermarks_s_yolov8_v1.pt')

# Load the image
image_path = "data-scraping/watermarked_images/watermarked_image_4.jpg"
image = Image.open(image_path)

# Perform object detection
results = model(image)

# Process and print results
for result in results:
    print(result)
```

```python
# Iterate through the detected boxes
for box in results[0].boxes:
    # Coordinates of the bounding box
    x1, y1, x2, y2 = box.xyxy[0]  # (top-left and bottom-right coordinates)
    # Confidence score
    confidence = box.conf[0]
    # Class name
    class_name = results[0].names[int(box.cls[0])]
    
    print(f"Detected {class_name} with confidence {confidence:.2f} at [{x1:.0f}, {y1:.0f}, {x2:.0f}, {y2:.0f}]")
```

```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw

# Load the image
image_path = "dataset/ai_image_detection/test_dataset/Authentic Images/A171.1.jpg"
image = Image.open(image_path)
draw = ImageDraw.Draw(image)

# Draw the bounding boxes on the image
for box in results[0].boxes:
    x1, y1, x2, y2 = box.xyxy[0]
    draw.rectangle([(x1, y1), (x2, y2)], outline="red", width=2)
    draw.text((x1, y1), f"{results[0].names[int(box.cls[0])]}: {box.conf[0]:.2f}", fill="red")

# Display the image
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis("off")
plt.show()
```
