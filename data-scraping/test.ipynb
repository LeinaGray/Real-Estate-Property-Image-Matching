```python
from selenium import webdriver
from selenium.webdriver.common.by import By

url = 'https://ohmyhome.com/en-ph/property-investments/results/?propertyType=House%20and%20Lot&selectedLang=PHL'

driver = webdriver.Chrome()
driver.get(url)
```

```python
import os, csv
```

```python
links = []
for i in range(44):
    driver.find_element(By.CLASS_NAME, "false").click()
    divs = driver.find_elements(By.CLASS_NAME, 'ProjectCardNoLogo_project-url__AqbeJ')
    for div in divs:
        a = div.find_element(By.TAG_NAME, "a")
        link = a.get_attribute('href')
        folder_path = "dataset"
        os.makedirs(folder_path, exist_ok=True)
        file_path = os.path.join(folder_path, "ohmyhome_links.csv")
        with open(file_path, 'a', encoding='utf-8', newline='') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow([link])
        links.append(link)
        print(link)
```
